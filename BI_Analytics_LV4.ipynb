## Task 8: Investment Duration
#### Objective: To Calculate the average investment duration.

```python
Duration_df = df[['Duration']]
print(Duration_df)
```

#### Looking at the above Column, it wont be possible to drectly find the mean (average) duration of the invstments. In order to get this done, we'd start by converting the duration ranges into numeric values. We can assume midpoints for the ranges.

```python
import pandas as pd
Duration_df = df[['Duration']]
mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
Duration_df["Duration_values"] = Duration_df["Duration"].map(mapping)
```

```python
Duration_df
```

```python
Average_duration = Duration_df["Duration_values"].mean()

print(f"\nThe average Investment duration of the participants is: {Average_duration} years")
```

## Task 10: Correlation Analysis
### Objective: To Explore potential correlations between factors.

### To perform correlation Analysis for the 'Age', 'Investment Duration', and 'Expected Returns', we have to ensure that these columns are perfectly and completely NUmeric columns, for the operation to go smoothly. We can achieve that by mapping the ranged items in the columns, with estimated figures (midpoints)

```python
df[["age", "Duration", "Expect"]]
mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
Duration_df["Duration_values"] = Duration_df["Duration"].map(mapping)

df[["age", "Duration", "Expect"]]
mapping = {
    '10%-20%': 15,
    '20%-30%': 25,
    '30%-40%': 35,
}
Expect_df["Expect_values"] = Expect_df["Expect"].map(mapping)
```

```python
df["duration_numeric"] = Duration_df["Duration_values"]
df["Expect_numeric"] = Expect_df["Expect_values"]
```

```python
df
```

### Hence, the Prepared data for the correlation analysis is presented as follows:

```python
df[["age", "duration_numeric", "Expect_numeric"]]
```

### Correlation Analysis

```python
Correlation_Matrix = df[["age", "duration_numeric", "Expect_numeric"]].corr()
Correlation_Matrix
```

### Using a ScatterPlot Diagram, and fitting a Regression line, to visualize the relationship between the Duration and Expectation, we have:

```python
# Scatter plot with regression line
sns.regplot(x=df["duration_numeric"], y=df["Expect_numeric"])
plt.title("Duration vs Returns Expectation")
plt.show()
```

## Analysis summary and Interpretation

#### The correlation analysis reveals that age has little to no impact on either duration or salary expectations. The correlation between age and duration (0.0518) indicates a very weak positive relationship, meaning that age does not significantly determine how long someone has worked. Similarly, the correlation between age and salary expectations (-0.0896) shows a very weak negative relationship, suggesting that as age increases, salary expectations might slightly decrease, but the effect is negligible. Overall, age does not play a significant role in shaping duration or salary expectations based on this dataset.

#### However, there is a weak positive correlation (0.2582) between duration and salary expectations, indicating that individuals with longer work experience tend to expect higher salaries. While this trend exists, the strength of the correlation suggests that other factors might also influence salary expectations. 

#### The scatter plot shows a **weak positive correlation** between **investment duration** and **returns expectation**, meaning longer durations are generally associated with higher expected returns. The **regression lines upward slope** suggests this trend, but the **widely spread data points** indicate that the relationship is not strong. The **shaded confidence interval** widens as duration increases, showing **greater uncertainty** in predicting returns for longer durations. Some **outliers** suggest that other factors, beyond duration, might influence returns expectations. Overall, while a **general upward trend exists**, the relationship remains **weak and inconsistent**.
